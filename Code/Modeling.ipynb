# Modeling with Doc2Vec

```python
import multiprocessing
cores = multiprocessing.cpu_count()
```

```python
train =df[['clean_text','Class']].iloc[:train_len]
test = df[['clean_text','Class']].iloc[train_len:]
train_other=df.drop(columns=['ID','Class','clean_text']).iloc[:train_len]
test_other=df.drop(columns=['ID','Class','clean_text']).iloc[train_len:]
```

```python
y_train =df[['Class']].iloc[:train_len]
y_test = df[['Class']].iloc[train_len:]
```

```python
train_tagged = train.apply(
    lambda r: TaggedDocument(words=tokenize_text(r['clean_text']), tags=[r.Class]), axis=1)
test_tagged = test.apply(
    lambda r: TaggedDocument(words=tokenize_text(r['clean_text']), tags=[r.Class]), axis=1)
```

```python
#instantiate our DM and DBOW models
d2v_model_dm = gensim.models.Doc2Vec(min_count=3, window=7, size=400, sample=1e-3, negative=5,dm=1, workers=2)
d2v_model_dbow = gensim.models.Doc2Vec(min_count=3, window=7, size=400, sample=1e-3, negative=5, dm=0, workers=2)
```

```python
d2v_model_dm.build_vocab(train_tagged)
```

```python
train_vecs_dm,test_vecs_dm=build_doc2vec(d2v_model_dm,train_tagged,test_tagged)
evaluate_features(train_vecs_dm,y_train,test_vecs_dm,y_test,RandomForestClassifier(max_depth=5, n_estimators=500,class_weight='balanced'))
```

```python
#Integrate the other features
X_train_dm=np.hstack((train_vecs_dm,train_other))
X_test_dm=np.hstack((test_vecs_dm,test_other))
```

```python
evaluate_features(X_train_dm,y_train,X_test_dm,y_test,RandomForestClassifier(max_depth=5, n_estimators=500,class_weight='balanced'))
```
